```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data: Creating a DataFrame with ages and genders
data = {
    'Age': [23, 45, 31, 22, 34, 29, 43, 35, 50, 27, 22, 33, 41, 29, 40],
    'Gender': ['Female', 'Male', 'Female', 'Female', 'Male',
               'Male', 'Female', 'Male', 'Female', 'Female',
               'Male', 'Female', 'Male', 'Male', 'Female']
}
df = pd.DataFrame(data)

# Set the style of seaborn
sns.set(style='whitegrid')

# Create a figure
plt.figure(figsize=(10, 6))

# Create a histogram for age distribution, separated by gender
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack', bins=10)

# Adding titles and labels
plt.title('Age Distribution by Gender', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data: Creating a DataFrame with ages and genders
data = {
    'Age': [23, 45, 31, 22, 34, 29, 43, 35, 50, 27,
            22, 33, 41, 29, 40, 38, 24, 31, 36, 50,
            60, 65, 70, 25, 28, 30, 37, 45, 52, 48,
            22, 36, 47, 54, 62, 29, 34, 41, 53, 58],
    'Gender': ['Female', 'Male', 'Female', 'Female', 'Male',
               'Male', 'Female', 'Male', 'Female', 'Female',
               'Male', 'Female', 'Male', 'Male', 'Female',
               'Female', 'Male', 'Female', 'Female', 'Male',
               'Male', 'Female', 'Female', 'Male', 'Female',
               'Male', 'Male', 'Female', 'Male', 'Male',
               'Female', 'Male', 'Female', 'Female', 'Male',
               'Male', 'Female', 'Male', 'Female', 'Female']
}

df = pd.DataFrame(data)

# Set the style of seaborn
sns.set(style='whitegrid')

# Create a figure with subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Histogram for age distribution by gender
sns.histplot(data=df, x='Age', hue='Gender', multiple='stack', bins=10, ax=axs[0])
axs[0].set_title('Age Distribution by Gender', fontsize=16)
axs[0].set_xlabel('Age', fontsize=14)
axs[0].set_ylabel('Count', fontsize=14)

# Box plot for age distribution by gender
sns.boxplot(data=df, x='Gender', y='Age', ax=axs[1])
axs[1].set_title('Age Distribution by Gender (Box Plot)', fontsize=16)
axs[1].set_xlabel('Gender', fontsize=14)
axs[1].set_ylabel('Age', fontsize=14)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
